```python
merged_data=df2.merge(df3)
print(merged_data)
```

```python
merged_data.head()
```

```python
merged_data=merged_data.rename(columns={'original_title':'title'})
print(merged_data)
```

```python
merged_data.head()
```

```python
FINAL_DATA=pd.merge(merged_data, df, on='title')
print(FINAL_DATA)
```

```python
FINAL_DATA.head()
```

# Number of movies per year

```python
FINAL_DATA['year'].value_counts()
```

# top10 title of the movies

```python
top10=FINAL_DATA.nlargest(10,'year')[['title','year']].set_index('title')
```

```python
print(top10)
```

# top5 lengthy movies and runtime_minutes

```python
top5_len=FINAL_DATA.nlargest(5,'runtime_minutes')[['title','runtime_minutes']].set_index('title')
```

```python
print(top5_len)
```

# Average rating of each movies

```python
FINAL_DATA.groupby('title')['averagerating'].mean()
```

# top10 genres and numvotes

```python
FINAL_DATA.nlargest(10,'numvotes')[['genres','numvotes']].set_index('genres')
```

# Highest average numvotes 

```python
FINAL_DATA.groupby('year')['numvotes'].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar('year','numvotes', data=FINAL_DATA)
plt.xlabel('year')
plt.ylabel('numvotes')
plt.title('Highest Average Numvotes')
plt.show()
```
